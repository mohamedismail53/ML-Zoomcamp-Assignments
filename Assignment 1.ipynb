## Q6 Solution

```python
Rolls_Royce = df[df['Make'] == 'Rolls-Royce']
Rolls_Royce.head()
```

```python
Rolls_Royce = Rolls_Royce[["Engine HP", "Engine Cylinders", "highway MPG"]]
Rolls_Royce
```

```python
X = Rolls_Royce.drop_duplicates()
```

```python
XTX = np.transpose(X).dot(X)
```

```python
inverse = np.linalg.inv(XTX)
inverse
```

```python
sum(sum(inverse))   # answer
```
